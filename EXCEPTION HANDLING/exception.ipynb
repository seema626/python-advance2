```python
def division2():
    try:
        num = float(input("Please enter numerator : "))
        den = float(input("Please enter denominator : "))
        div = num/den
        print(f"Numerator : {num}, Denominator : {den} , Division : {div:.4f}")
        return div
    except Exception as e:
        print(f"Exception Occured : {e}")
```

### Error handlling for different exceptions

1.zero Division error
2.value errors

```python
def division3():
    try:
        num = float(input("Please enter numerator : "))
        den = float(input("Please enter denominator : "))
        div = num/den
        print(f"Numerator : {num}, Denominator : {den} , Division : {div:.4f}")
        return div
    except ZeroDivisionError:
        print("Denominator in division cannot be Zero")
    except ValueError:
        print("Numerator and Denominator should both be numbers, blank input is not allowed")
```

```python
# Num - 10, Den - 3
d1 = division3()
print("done")
```

```python
d1
```

```python
# Num - 20, Den - 0
d2 = division3()
print("done")
```

```python
# Num - d
d3 = division3()
print("done")
```

```python
# Num - blank
d3 = division3()
print("done")
```

#### 3. Aggreation operations in numpy

```python
# find out average marks inside the list
marks = [83, 72, 67, 89, 93, 70, 52, 35]
```

```python
# initialize sum to 0
s = 0
```

```python
# Apply for loop on marks
for i in marks:
    s = s + i
```

```python
# find out number of student
count = len(marks)
```

```python
# Average marks
avg = s/count
```

```python
# print average marks
print(f"Average of {marks} : {avg:.2f}")
```

```python
avg1 = np.mean(marks)
avg1
```
